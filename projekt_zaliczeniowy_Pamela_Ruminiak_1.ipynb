```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
#angielski stopwords
stop = set(stopwords.words("english"))
wegiel["text_without_stopwords"] = wegiel["text"].apply(lambda x: " ".join([elem for elem in x.split() if elem not in (stop)]))
```

```python
#polski stopwords
with open("stopwords.txt", "r") as f:
 new_stopwords = []
 for line in f.readline():
    new_stopwords.append(line)

new_stopwords_set = set(new_stopwords)
polskie_stopwords = [elem.replace("\n", "") for elem in new_stopwords_set]
wegiel["text_without_stopwords_polskie"] = wegiel["text"].apply(lambda x: " ".join([elem for elem in x.split() if elem.lower() not in (polskie_stopwords)]))
```
